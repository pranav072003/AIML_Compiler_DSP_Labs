```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
N=500
norm = np.random.normal(1,1,N);
#print(norm)
plt.hist(norm, bins=20, edgecolor='blue', color='yellow')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Frequency distribution for a normal distribution')
```

```python
#draw box plot for normal distribution
Q1 = np.percentile(norm,25)
Q3 = np.percentile(norm,75)
plt.boxplot(norm, vert=False)
plt.plot(Q1, 1, 'rx')
plt.plot(Q3, 1, 'rx')
IQR = Q3-Q1;
dis1 = np.abs(max(norm) - Q3);
dis2 = np.abs(min(norm) - Q1);
#check for extreme outliers
a = (dis1>3*IQR)
b = (dis2>3*IQR)
#check for mild outliers
c = (dis1<=3*IQR and dis1>1.5*IQR)
d = (dis2<=3*IQR and dis2>1.5*IQR)
if a or b:
    print('Data has extreme outliers');
elif c or d:
    print('Data has mild outliers');
else:
    print('Data has no outliers')
```

```python
#repeat for exponential distribution
expo = np.random.exponential(1,N)
# print(expo)
plt.subplot(1,2,1)
a = plt.hist(expo, bins=40, edgecolor='green', color='magenta')  #cumulative histogram for bins
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Frequency distribution for an exponential distribution')

# creating relative prob distribution
b = a[0] / N;
bin_edges = a[1];
bin_midpoints = (bin_edges[1:] + bin_edges[:-1]) / 2
plt.subplot(1,2,2)
plt.plot(bin_midpoints, b)
```
