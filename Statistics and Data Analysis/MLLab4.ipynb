```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 25, 1)
x

pmf = poisson.pmf(x, mu=10)
print(pmf.dtype)
pmf = np.round(pmf, 5)

plt.subplot(1, 2, 1)

plt.bar(x, pmf)
plt.xlabel("Data Values-->")
plt.ylabel("Probability-->")
plt.title("LAMBDA- 15")

plt.subplot(1, 2, 2)

pmf = poisson.pmf(x, mu=1)
pmf = np.round(pmf, 5)

plt.bar(x, pmf)
plt.xlabel("Data Values-->")
plt.ylabel("Probability-->")
plt.title("LAMBDA- 1")
```

```python
# creating poisson dist by hard coding
import math
x = np.arange(0, 20, 1)

def poisson_dist(x, mu):
  poiss = []
  u = np.exp(-1*mu)
  for i in x:
    # print(mu**i)
    poiss.append((mu**i)*u/math.factorial(i))
  return poiss

pmf = poisson_dist(x, mu=10)
plt.bar(x, pmf)
# works till values of x till 1000, quite strange compared to limit of 25 in case of scipy
```
